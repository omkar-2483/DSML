Perform the following operations using Python on the data set
House_Price Prediction dataset. Compute standard deviation, variance and
percentiles using separate commands, for each feature. Create a histogram
for each feature in the dataset to illustrate the feature distributions.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('House_Price.csv')
```

```python
df.head()
```

```python
numerical_columns = df.select_dtypes(['float64', 'int64']).columns

for column in numerical_columns:
    print('for ', column)
    print('Mean: ',df[column].mean())
    print('Max value : ',df[column].max())
    print('Min value : ',df[column].min())
    print('Std deviation: ',df[column].std())
    print('Variance: ',df[column].var())
    print('25% : ',df[column].quantile(0.25))
    print('50% : ',df[column].median())
    print('75% : ',df[column].quantile(0.75))
    print()
```

```python
def plot_hist(column):
    sns.histplot(df[column], kde=True, bins=20)
for column in numerical_columns:
    plot_hist(column)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
   
```

```python
# Create histograms for each feature in the dataset
df.hist(bins=20, figsize=(15, 12))
plt.suptitle('Histograms of Features')
plt.tight_layout()
plt.show()
```
