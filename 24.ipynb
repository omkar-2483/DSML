Perform the following operations using Python on a suitable data set,
counting unique values of data, format of each column, converting variable
data type (e.g. from long to short, vice versa), identifying missing values
and filling in the missing values.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('House_Price.csv')
df.head()
```

```python
# count unique values
print('unique value count: \n')
airport_count = df['airport'].value_counts()
waterbody_count = df['waterbody'].value_counts()
bus_ter_count =df['bus_ter'].value_counts()

print(airport_count,'\n', waterbody_count, '\n' , bus_ter_count)
```

```python
# Check the data types of each column
data_types = df.dtypes
print("Data types of each column:\n", data_types)
```

```python
# convert data types
df['room_num'] = df['room_num'].astype('int64')

df['age'] =df['age'].astype('int64')

df['rainfall'] = df['rainfall'].astype('float64')

df.head()
```

```python
# Identify missing values (NaN) in each column
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# fill missing n_hos_beds with mean
df['n_hos_beds']=df['n_hos_beds'].fillna(df['n_hos_beds'].mean())
```

```python
#fill waterbody with mode
df['waterbody'].mode()[0]
df['waterbody'] = df['waterbody'].fillna(df['waterbody'].mode()[0])
```
