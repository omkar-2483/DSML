```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Customer-Churn.csv')
# df.head()
```

```python
# remove useless columns
df= df.drop(columns=['customerID'])
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.boxenplot(x=df['TotalCharges'])
```

```python
df['TotalCharges'].describe()
```

```python
percentile_99 = df['TotalCharges'].quantile(0.99)
df[df['TotalCharges']>percentile_99]
```

```python
df['TotalCharges']= df['TotalCharges'].apply(lambda x: percentile_99 if x>percentile_99 else x)
```

Variable Transformation

```python
# convert target variable
df['Churn']= df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df.head()
```

```python
# dummy incoding for all categorical variables
df = pd.get_dummies(df).astype(int)
```
