13. Use the covid_vaccine_statewise.csv dataset and perform the following
analytics.


a. Describe the dataset

b. Number of persons state wise vaccinated for first dose in India

c. Number of persons state wise vaccinated for second dose in India

```python
import pandas as pd
```

```python
df = pd.read_csv("Covid_Vaccine_Statewise.csv")
df.head()
```

```python
states = df['State'].unique()
states= states[1:]
states
```

```python
print('Number of persons state wise vaccinated for first dose in India')
for state in states:
    doses = df[df['State'] == state]['First Dose Administered'].sum()
    print(f'{state}: {doses}')
```

```python
print('Number of persons state wise vaccinated for second dose in India')
for state in states:
    doses = df[df['State'] == state]['Second Dose Administered'].sum()
    print(f'{state}: {doses}')
```
