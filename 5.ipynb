Write a program to do: A dataset collected in a cosmetics shop showing
details of customers and whether or not they responded to a special offer
to buy a new lip-stick is shown in table below. (Use library commands)
According to the decision tree you have made from the previous training
data set, what is the decision for the test data: [Age < 21, Income = Low,
Gender = Female, Marital Status = Married]?

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('cosmetics_shop_data.csv')
df.head()
```

```python
df['Buys']= df['Buys'].apply(lambda x: 1 if x=='Yes' else 0)
df=pd.get_dummies(df).astype(int)
```

```python
df.head()
```

```python
# x-Y split 
X= df.drop('Buys', axis=1)
y=df['Buys']
```

```python
clf= DecisionTreeClassifier(criterion='entropy', random_state=0)
clf.fit(X,y)
```

```python
input =[{'Age': 19, 'Income': 'Low', 'Gender': 'Female', 'Marital Status': 'Married'}]
input = pd.DataFrame(input)
input = pd.get_dummies(input).astype(int)
print(input)

# Align columns with training data (add missing columns as 0)
input = input.reindex(columns=df.columns.drop('Buys'), fill_value=0)
print(input)

prediction = clf.predict(input)[0]

print(prediction)
```

```python
if(prediction==1):
    print('customer will buy cosmetic')
else:
    print('customer will not buy cosmetic')
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y, clf.predict(X)))
```
