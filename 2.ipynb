Perform the following operations using Python on the Telecom_Churn
dataset. Compute and display summary statistics for each feature available
in the dataset using separate commands for each statistic. (e.g. minimum
value, maximum value, mean, range, standard deviation, variance and
percentiles).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Customer-Churn.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.sort_values(by='TotalCharges')
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
print('Mean of tenure: ',df['tenure'].mean())
print('Max value of tenure: ',df['tenure'].max())
print('Min value of tenure: ',df['tenure'].min())
print('Std deviation of tenure: ',df['tenure'].std())
print('Variance of tenure: ',df['tenure'].var())
print('25% of tenure: ',df['tenure'].quantile(0.25))
print('50% of tenure: ',df['tenure'].median())
print('75% of tenure: ',df['tenure'].quantile(0.75))
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

print("Summary Statistics for Numerical Features:\n")
for column in numerical_columns:
    if(column=='SeniorCitizen'): continue
    print(f"Feature: {column}")
    print(f"Minimum Value: {df[column].min()}")
    print(f"Maximum Value: {df[column].max()}")
    print(f"Mean Value: {df[column].mean()}")
    print(f"Range: {df[column].max() - df[column].min()}")
    print(f"Standard Deviation: {df[column].std()}")
    print(f"Variance: {df[column].var()}")
    print(f"25th Percentile: {df[column].quantile(0.25)}")
    print(f"50th Percentile (Median): {df[column].median()}")
    print(f"75th Percentile: {df[column].quantile(0.75)}")
    print("-" * 40)
```

```python
df.drop(columns=['customerID'], inplace=True)

# for categorical variables
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    print(f"mode of {col}: {df[col].mode()}")
```
