Use Iris flower dataset and perform following :
1. List down the features and their types (e.g., numeric, nominal)
available in the dataset. 2. Create a histogram for each feature in the
dataset to illustrate the feature distributions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv', header=0)
df=df.drop(columns=['Id'])
df.head()
```

```python
# List of the features and their types
features = df.columns[:-1]  # Exclude the target column (assumed to be the last column)
feature_types = df.dtypes

print("Features and their types:")
for feature, dtype in zip(features, feature_types):
    print(f"{feature}: {'Numeric' if dtype == 'float64' or dtype == 'int64' else 'Nominal'}")
```

```python
# histograms for each feature
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)  # Arrange subplots in a 2x2 grid
    sns.histplot(df[feature], kde=True, bins=15, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

12.Use Iris flower dataset and perform following :
1. Create a box plot for each feature in the dataset.
2. Identify and discuss distributions and identify outliers from them.

```python
# Create box plots for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features): 
    plt.subplot(2, 2, i + 1)  # Arrange subplots in a 2x2 grid
    sns.boxplot(x=df[feature], color='skyblue')
    plt.title(f'Box Plot of {feature}')
    plt.xlabel(feature)

plt.tight_layout()
plt.show()
```

Circular dots outside the box but within the whiskers: These are considered mild outliers.
Circular dots well outside the whiskers: These could be extreme outliers.
